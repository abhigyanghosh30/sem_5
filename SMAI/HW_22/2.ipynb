```python
import numpy as np
from matplotlib import pyplot as plt

x_1 = np.random.uniform(-100,100,500)
x_2 = np.random.uniform(-100,100,500)
x = np.hstack((x_1,x_2))
y_1 = np.ones(500) * x_1
y_2 = np.ones(500) * x_2 * -1
y = np.hstack((y_1,y_2))
noise = np.random.normal(0,10,1000)
y = y + noise
X = np.vstack((x,y)).T
np.random.shuffle(X)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
X_1 = X[:500]
X_2 = X[500:]
for i in range(100):
    line1 = np.polyfit(X_1[:,0],X_1[:,1],1)    
    line2 = np.polyfit(X_2[:,0],X_2[:,1],1)
    X_1 = []
    X_2 = []
    for j in range(X.shape[0]):
        d1 = np.abs(X[j,0]*line1[0] - X[j,1] + line1[1])/(line1[0]**2 + 1)
        d2 = np.abs(X[j,0]*line2[0] - X[j,1] + line2[1])/(line2[0]**2 + 1)
        if(d1<d2):
            X_1.append(X[j])
        else:
            X_2.append(X[j])
    X_1 = np.array(X_1)
    X_2 = np.array(X_2)

plt.scatter(X_1[:,0],X_1[:,1],c='r')
plt.scatter(X_2[:,0],X_2[:,1],c='b')
```
