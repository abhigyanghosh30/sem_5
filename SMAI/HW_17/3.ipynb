```python
from matplotlib import pyplot as plt
import numpy as np
a = [0.6,0.4,0.3,-0.2,-0.3]
x = [-2,-1,0,1,2]

for i in range(3000):
    l = len(x)
    last_x = x[l-5:l]
    new_x = np.dot(a,last_x)
    x.append(new_x)

x = np.array(x)
x = x + np.random.normal(0,1,(3005)) 
plt.plot([i for i in range(len(x))],x)
```

```python
def error(w):
    d = np.shape(w)[0]
    sum = 0
    for i in range(d+1,len(x)):
        x_new = x[i-d:i]
        pred = np.dot(w,x_new)
        sum += x[i+1]-pred
    return sum*sum
```
