```python
from matplotlib import pyplot as plt
import numpy as np
a = [0.7,0.3,0.4,-0.2,-0.6]
x = [-3,-2,-1,0,1]

for i in range(1000):
    l = len(x)
    last_x = x[l-5:l]
    new_x = np.dot(a,last_x)
    x.append(new_x)

x = np.array(x)
x = x + np.random.normal(0,0.5,(1005)) 
plt.scatter([i for i in range(len(x))],x)
```

```python
def error(w):
    d = np.shape(w)[0]
    sum = 0
    for i in range(len(x)-d):
        x_new = x[i:i+d]
        pred = np.dot(w,x_new)
        sum += x[i+1]-pred
    return sum*sum
```

```python
error(w=np.array([0.1]))
```
