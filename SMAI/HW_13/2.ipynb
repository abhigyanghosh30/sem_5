```python
import numpy as np
import csv

data = []
labels = []

with open('sample_train.csv','r') as f:
    rows = csv.reader(f)
    for row in rows:
        data.append(row[1:])
        labels.append(row[0])
data = np.matrix(data).astype(np.float)
data_mean = np.mean(data,axis=0)
data_cov = np.linalg.inv(np.cov(data))
data = data - data_mean
data = np.dot(data_cov,data) 
labels = np.array(labels).astype(np.float)
```

```python
its = 0

def sigmoid_activation(w,n=1.5,gamma = 10):
    print(w)
    
    global its
    
    its += 1
    if(its>100):
        return w
    
    y = []
    for i in range(len(labels)):
        y.append((labels[i]+1)/2)
    
    z = gamma * np.matmul(data,w)
    
    h = (np.exp(-z)))
    y_10 = np.empty((6000,10))
    for i in range(10):
        y_10[:,i]=y
    gradient = np.matmul(np.transpose(data),h-y_10) / len(data)
    w = w - n * gradient
    return sigmoid_activation(w)
```
